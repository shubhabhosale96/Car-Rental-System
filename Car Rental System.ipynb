```python
cars ={
    "car1":{'id':'C001','brand':'Maruti','Model':'Swift','Available':'Yes','hourly_rate':200},
    "car2":{'id':'C002','brand':'Ford','Model':'Explorar','Available':'Yes','hourly_rate':400},
    "car3":{'id':'C003','brand':'Hondai','Model':'Creta','Available':'Yes','hourly_rate':600},
    "car4":{'id':'C004','brand':'Maruti','Model':'Ertiga','Available':'yes','hourly_rate':800}
    
}
customer_rentals = []
earnings = 0

for i,j in cars.items():
    print("Car Rental Shop")
    print("1. Display Available Cars")
    print("2. Rent a Car")
    print("3. Return a Car")
    print("4. Display Earnings")
    print("5. Exit")
    choice = input("Enter your choice (1/2/3/4): ")
    
    if choice == "1":
        print("\nAvailable Cars:")
        for car, info in cars.items():
            if info["Available"]:
                print(f"{car} ({info['Model']}) - Hourly Rate: {info['hourly_rate']} Rs")
    elif choice == "2":
        print("\nAvailable Cars:")
        for car, info in cars.items():
            if info["Available"]:
                print(f"{car} ({info['Model']}) - Hourly Rate: {info['hourly_rate']} Rs")
  
        car_choice = input("Enter the car you want to rent: ")
        if car_choice in cars and cars[car_choice]["Available"]:
            customer_name = input("Enter your name: ")
            rental_hours = int(input("Enter the number of hours you want to rent the car: "))
            if rental_hours <= 0:
                print("Invalid rental duration. Please enter a valid duration.")
                continue

            bill_amount = cars[car_choice]["hourly_rate"] * rental_hours
            print(f"Rented {car_choice} for {rental_hours} hours. Your bill is {bill_amount}")
            cars[car_choice]["Available"] = False
            customer_rentals.append({
                "car": car_choice,
                "customer_name": customer_name,
                "rental_hours": rental_hours,
                "bill_amount": bill_amount,
            })
            earnings += bill_amount
        else:
            print("Sorry, the selected car is not available.")
    elif choice == "3":
        if not customer_rentals:
            print("No cars have been rented yet.")
        else:
            print("\nCars Currently Rented:")
            for index, rental in enumerate(customer_rentals, 1):
                print(f"{index}. {rental['customer_name']} rented {rental['car']} for {rental['rental_hours']} hours")

            return_choice = input("Enter the number of the car you are returning: ")
            return_choice = int(return_choice) - 1

            if 0 <= return_choice < len(customer_rentals):
                returned_car = customer_rentals.pop(return_choice)
                cars[returned_car["car"]]["available"] = True
                print(f"Thank you, {returned_car['customer_name']}! You have returned {returned_car['car']}.")
            else:
                print("Invalid choice. Please enter a valid car number.")
    elif choice == "4":
        print(f"Total Earnings: {earnings} Rs")
      
    elif choice == "5":
        break

    else:
        print("Invalid choice. Please select a valid option.")
            
```
